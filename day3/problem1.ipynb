```python
import numpy as np
```

```python
# get list of rotations
input_array = np.loadtxt("inputDay3.txt", dtype=str)

# numpy array to list
input_list = input_array.tolist()
```

```python
# Set initial count
count = 0

# Loop over inpus
for inp in input_list:

    # Loop over characters
    temp_number = inp[0]
    index_char = 0
    for i, char in enumerate(inp[1:-1]):

        # Check if number is larger than largest known
        if int(char) > int(temp_number):
            temp_number = char
            index_char = i+1

    # set initial final number
    final_number = temp_number

    # loop for a second time
    temp_number = inp[index_char+1]
    for char in inp[index_char+2:]:

        # Check if number is larger than largest known
        if int(char) > int(temp_number):
            temp_number = char

    # set final number
    final_number = final_number+temp_number

    # adjust count
    count += int(final_number)

print(f'Final solution for this problem is: {count}')
```
