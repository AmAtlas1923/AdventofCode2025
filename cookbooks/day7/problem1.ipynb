```python
# get list of matrices
input = [line.strip() for line in open("inputDay7.txt")]

# write to a matrix
input = [list(row) for row in input]
```

```python
# set counter
counter = 0

# Find start index and remove first row
locations = set([input[0].index('S')])
inp = input[1:]

# loop over rows
for line in inp:

    # loop over locations
    for loc in locations.copy():

        # check for a split
        if line[loc] == '^':

            # add to counter
            counter += 1

            # remove location
            locations.remove(loc)

            # add new locations
            locations.update({loc - 1, loc + 1})

print(counter)
```
