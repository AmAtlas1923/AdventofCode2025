```python
import numpy as np
```

```python
# get list of matrices
input = np.loadtxt('inputDay7.txt', dtype=str)

input = np.array([list(row) for row in input])
```

```python
# Find start index and remove first row
index_S = input[0,:].tolist().index('S')
locations = set([index_S])
paths = [1 if i == index_S else 0 for i in range(len(input[0, :]))]
inp = input[1:,:]

# loop over rows
for line in inp:

    paths_new = [0] * len(input[0,:])
    for loc in set(locations):
        if line[loc] == '^':

            # remove location
            locations.remove(loc)

            # add new locations
            locations.update({loc - 1, loc + 1})

            paths_new[loc-1] += paths[loc]
            paths_new[loc+1] += paths[loc]
            
        else:
            paths_new[loc] += paths[loc]

    paths = paths_new

print(int(np.sum(paths)))
```
