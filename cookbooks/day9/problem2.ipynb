```python
from shapely.geometry import Polygon, box
```

```python
# get list of coordinates
input = [line.strip() for line in open("example.txt")]

# get matrix of coordinates
input = [line.split(',') for line in input]

# set to int
input = [[int(x) for x in lis] for lis in input]

print(input)
```

```python
opps = []
for i in range(len(input)):
    for j in range(i+1,len(input)):
        opp = (abs(input[i][0]-input[j][0])+1)*(abs(input[i][1]-input[j][1])+1)
        opps.append((opp,i,j))

opps = sorted(opps, key=lambda opp: opp[0], reverse=True)

# create polygon
polygon = Polygon(input)

for opp in opps:
    coord1 = input[opp[1]]
    coord2 = input[opp[2]]
    x_min, x_max = sorted([coord1[1], coord2[1]])
    y_min, y_max = sorted([coord1[0], coord2[0]])
    
    if polygon.covers(box(y_min,x_min,y_max,x_max)):
        print(opp[0])
        break
```

```python
opps = []
for i in range(len(input)):
    for j in range(i+1,len(input)):
        opp = (abs(input[i][0]-input[j][0])+1)*(abs(input[i][1]-input[j][1])+1)
        opps.append((opp,i,j))

opps = sorted(opps, key=lambda opp: opp[0], reverse=True)

n_input = len(input)

print(opps)

for opp, i,j in opps:

    coord1 = input[i]
    coord2 = input[j]
    x_min, x_max = sorted([coord1[1], coord2[1]])
    y_min, y_max = sorted([coord1[0], coord2[0]])

    upper = [(y_min,x) for x in range(x_min,x_max+1)]
    lower = [(y_max,x) for x in range(x_min,x_max+1)]
    left = [(y,x_min) for y in range(y_min,y_max+1)]
    right = [(y,x_max) for y in range(y_min,y_max+1)]
    points = set(upper + lower + left + right)

    all_inside = True
    for point in points:

        inside = False
        # loop over polygon
        for inp in range(len(input)):
            py, px = point[0], point[1]
            y1, x1 = input[inp]
            y2, x2 = input[(inp + 1) % n_input]

            if ((y1 > py) != (y2 > py)):
                if px < x1:
                    inside = not inside

        if not inside:
            all_inside = False
            break
        
    if all_inside:
        print(opp,input[i], input[j], i,j)
        break
```
