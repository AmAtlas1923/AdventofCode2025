```python
# get list of coordinates
input = [line.strip() for line in open("inputDay9.txt")]

# get matrix of coordinates
input = [line.split(',') for line in input]

# set to int
input = [[int(x) for x in lis] for lis in input]
```

```python
counter = 0

for i in range(len(input)):
    for j in range(i+1,len(input)):

        opp = (abs(input[i][0]-input[j][0])+1)*(abs(input[i][1]-input[j][1])+1)

        counter = max(counter,opp)

print(counter)
```
