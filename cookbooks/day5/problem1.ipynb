```python
import numpy as np
```

```python
# Get input of ranges and numbers
input = np.genfromtxt("inputDay5.txt", dtype=str, comments=None, delimiter="\n", filling_values="")

# mask the ranges
mask = np.char.find(input, '-') >= 0

# set ranges and numbers
ranges = input[mask]
numbers = input[~mask]
```

```python
# set range numbers and numbers to int
range_numbers = [tuple(map(int, r.split('-'))) for r in ranges]
numbers = [int(n) for n in numbers]

# set initial counter
counter = 0

# loop over numbers and ranges
for number in numbers:
    for range_n in range_numbers:

        # check conditions and break
        if (number >= range_n[0]) & (number <= range_n[1]):
            counter += 1
            break

# print result
print(f"Following solution found: {counter}")
```
