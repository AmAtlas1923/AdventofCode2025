```python
# split string function
def split_string(x,n):
    return [x[i:i+n] for i in range(0, len(x), n)]

# find divisors
def find_divisors(n):
    return [x for x in range(1,n+1) if n%x==0]
```

```python
# get list of string ranges
input_text = open("inputDay2.txt").readline()

# get ranges in list
input_ranges = input_text.split(',')
```

```python
# set initial count
count = 0

# loop over ranges
for range_n in input_ranges:

    # split by -
    range_start_end = range_n.split('-')

    # set ranges
    range_list = list(range(int(range_start_end[0]), int(range_start_end[1]) + 1))

    # give the length of the id
    range_list_str = [str(s) for s in range_list]
    range_list_char_count = [len(s) for s in range_list_str]

    # give a list of divisors
    divisor_list = [[n for n in find_divisors(x) if  n != x] for x in range_list_char_count]

    # filter all numbers that have len 0
    index_count_list = [len(x) for x in divisor_list]
    rlist, dlist, ilist = [], [], []
    for r,d,i in zip(range_list_str, divisor_list, index_count_list):
        if i > 0:
            ilist.append(i)
            rlist.append(r)
            dlist.append(d)

    # loop over numbers
    for i, r in enumerate(rlist):
        for d in dlist[i]:
            splitted_string = split_string(r,d)
            count_dist = len(set(splitted_string))
            if count_dist == 1:
                count += int(r)
                break

print(f'Final solution for this problem is: {count}')
```
