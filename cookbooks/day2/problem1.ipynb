```python
# get list of string ranges
input_text = open("inputDay2.txt").readline()

# get ranges in list
input_ranges = input_text.split(',')
```

```python
# set initial count
count = 0

# loop over ranges
for range_n in input_ranges:

    # split by -
    range_start_end = range_n.split('-')
    
    # set ranges
    range_list = list(range(int(range_start_end[0]), int(range_start_end[1]) + 1))

    # give the length of the id
    range_list_str = [str(s) for s in range_list]
    range_list_char_count = [len(s) for s in range_list_str]

    # get the numbers that can be divided by two
    even_list = [x%2 for x in range_list_char_count]
    rlist, clist = [], []
    for r,c,e in zip(range_list_str, range_list_char_count, even_list):
        if e == 0:
            rlist.append(r)
            clist.append(c)

    # get divisions
    for i, r in enumerate(rlist):

        # split string
        r_start = r[:int(clist[i]/2)]
        r_end = r[int(clist[i]/2):]

        # check number
        if r_start == r_end:
            count += int(r)

print(f'Final solution for this problem is: {count}')

```
