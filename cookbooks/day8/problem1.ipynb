```python
# get list of coordinates
input = [line.strip() for line in open("inputDay8.txt")]

# get matrix of coordinates
input = [line.split(',') for line in input]

# set to int
input = [[int(x) for x in lis] for lis in input]
```

```python
# Credits https://www.geeksforgeeks.org/dsa/introduction-to-disjoint-set-data-structure-or-union-find-algorithm/
class UnionFind:
    def __init__(self, size):
      
        # Initialize parent
        self.parent = list(range(size))
    
    def find(self, i):
      
        # recursively find parent
        if self.parent[i] == i:
            return i
        return self.find(self.parent[i])
    
    def unite(self, i, j):
      
        # Representative of set containing i
        irep = self.find(i)
        
        # Representative of set containing j
        jrep = self.find(j)
        
        # adjust parent
        self.parent[irep] = jrep
```

```python
# initialise inputs
list_pairs_distances = []
n = len(input)

# double loop over coordinates
for i in range(n):
    for j in range(i+1,n):

        # Calculate distances
        list_pairs_distances.append((((input[i][0]-input[j][0])**2 + (input[i][1]-input[j][1])**2 + (input[i][2]-input[j][2])**2), i, j))

# sort distances
list_pairs_distances = sorted(list_pairs_distances)

# inialize union find class
uf = UnionFind(n)

# loop over 1000 smallest distances
for idx in range(1000):

    # get indices from ranked list
    _, i, j = list_pairs_distances[idx]

    # unite to update parents
    uf.unite(i, j)

# find number of children to determine largest circuits
circuits = [0] * n
for i in range(n):
    j = uf.find(i)
    circuits[j-1] += 1

# prod 3 largest circuits
circuits = sorted(circuits)
print(circuits[-1]*circuits[-2]*circuits[-3])
```
