```python
import numpy as np
import math
```

```python
# get list of coordinates
input = np.loadtxt('inputDay8.txt', dtype=str)

# get matrix of coordinates
input = np.char.split(input, ',').tolist()

# set to int
input = [[int(x) for x in lis] for lis in input]
```

```python
# Credits https://www.geeksforgeeks.org/dsa/introduction-to-disjoint-set-data-structure-or-union-find-algorithm/
class UnionFind:
    def __init__(self, size):
      
        # Initialize parent
        self.parent = list(range(size))
    
    def find(self, i):
      
        # recursively find parent
        if self.parent[i] == i:
            return i
        return self.find(self.parent[i])
    
    def unite(self, i, j):
      
        # Representative of set containing i
        irep = self.find(i)
        
        # Representative of set containing j
        jrep = self.find(j)
        
        # adjust parent
        self.parent[irep] = jrep
```

```python
# initialise inputs
list_pairs_distances = []
n = len(input)

# double loop over coordinates
for i in range(n):
    for j in range(i+1,n):

        # Calculate distances
        list_pairs_distances.append(((math.sqrt((input[i][0]-input[j][0])**2 + (input[i][1]-input[j][1])**2 + (input[i][2]-input[j][2])**2)), i, j))

# sort distances
list_pairs_distances = sorted(list_pairs_distances)

# inialize union find class
uf = UnionFind(n)

# loop over all distances ti find parents
for idx in range(len(list_pairs_distances)):

    # get indices from ranked list
    _, i, j = list_pairs_distances[idx]

    # unite to update parents
    uf.unite(i, j)
    
    # Check if all parents give 1 set
    if len(set([uf.find(k) for k in range(n)])) == 1:
        print(input[i][0]*input[j][0])
        break
```
