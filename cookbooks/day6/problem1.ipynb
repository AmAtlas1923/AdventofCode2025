```python
import numpy as np
```

```python
# get list of matrices
input = np.loadtxt('inputDay6.txt', dtype=str)
```

```python
# input = np.array([["123", "328",  "51", "64"], ["45", "64",  "387", "23"], ["6", "98",  "215", "314"], ["*", "+", "*", "+"]])

numbers = input[:-1,:].astype(int)
signs = input[-1,:]

counter = 0

for i, s in enumerate(signs):

    col = numbers[:,i]

    if s == '*':
        counter += np.prod(col)

    elif s == '+':
        counter += np.sum(col)

print(counter)
```
