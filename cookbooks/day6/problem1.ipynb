```python
# read the lines
lines = open('inputDay6.txt').read().splitlines()

# find separators
seps = [i for i,char in enumerate(lines[-1]+' *') if char != ' ']
n_seps = len(seps)

# loop over lines to create input
input = []
for i in range(len(lines)):
    input_line = []
    for j in range(n_seps-1):
        input_line.append(lines[i][seps[j]:seps[j+1]-1])
    if i == len(lines)-1:
        input_line = [j.strip(' ') for j in input_line]
    input.append(input_line)
```

```python
numbers = [list(map(int, row)) for row in input[:-1]]
signs = input[-1]

counter = 0

for i, s in enumerate(signs):

    col = [row[i] for row in numbers]

    if s == '*':
        prod = 1
        for n in col:
            prod *= n
        counter += prod

    elif s == '+':
        counter += sum(col)

print(counter)
```
