```python
import numpy as np
```

```python
# read the lines
lines = open('inputDay6.txt').read().splitlines()

# find separators
seps = [i for i,char in enumerate(lines[-1]+' *') if char != ' ']
n_seps = len(seps)

# loop over lines to create input
input = []
for i in range(len(lines)):
    input_line = []
    for j in range(n_seps-1):
        input_line.append(lines[i][seps[j]:seps[j+1]-1])
    if i == len(lines)-1:
        input_line = [j.strip(' ') for j in input_line]
    input.append(input_line)

input = np.array(input)
```

```python
numbers = input[:-1,:]
signs = input[-1,:]
counter = 0

# get length longest number in col
for i, s in enumerate(signs):

    max_length = len(max(numbers[:,i],key=len))

    temp_array = np.array([list(j.ljust(max_length)) for j in numbers[:,i]])
    number_list = []
    for j in range(max_length):
        number_list.append(int(''.join(temp_array[:,j])))

    if s == '*':
        counter += np.prod(np.array(number_list))

    elif s == '+':
        counter += np.sum(np.array(number_list))

print(counter)
```
