```python
import numpy as np
import math
```

```python
# get list of rotations
array_rotations = np.loadtxt("inputDay1.txt", dtype=str)

# numpy array to list
list_rotations = array_rotations.tolist()
```

```python
# split list numbers
list_numbers = [int(x.replace("L","-").replace("R","")) for x in list_rotations]

# start number
current_number = 50

# zero counter
counter = 0

for number in list_numbers:

    # get the sign
    sign = math.copysign(1,number)

    # rotation number
    N_rot, __ = divmod(abs(number),100)
    counter += N_rot

    # correct for rotations
    number -= sign*N_rot*100

    # check if passes 0
    if ((current_number+number >= 100) | (current_number+number <= 0)) & (current_number != 0):
        counter += 1

    # Add rotation
    current_number = (current_number+number)%100

# print result
print(f"Following solution found: {counter}")

```
