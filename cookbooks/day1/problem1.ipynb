```python
# get list of rotations
list_rotations = [line.strip() for line in open("inputDay1.txt")]
```

```python
# split list numbers
list_numbers = [int(x.replace("L","-").replace("R","")) for x in list_rotations]

# start number
current_number = 50

# zero counter
counter = 0

for number in list_numbers:

    # Add rotation
    current_number = (current_number+number)%100

    # add to counter when 0
    if current_number == 0:
        counter += 1

# print result
print(f"Following solution found: {counter}")

```
