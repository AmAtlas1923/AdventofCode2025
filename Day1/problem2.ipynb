```python
import numpy as np
```

```python
# get list of rotations
array_rotations = np.loadtxt("inputDay1Problem1.txt", dtype=str)

# numpy array to list
list_rotations = array_rotations.tolist()
```

```python
# split list numbers
list_numbers = [int(x[1:]) for x in list_rotations]
list_directions = [x[0] for x in list_rotations]

# start number
current_number = 50
prev_number = 50

# zero counter
counter = 0

for i, number in enumerate(list_numbers):

    # count rotations
    calc_number = number%100

    # rotation number
    N_rot = int((number-calc_number)/100)
    counter += N_rot

    # left click
    if list_directions[i] == 'L':

        # subtract number
        current_number -= calc_number

        # correct for number out of range
        if current_number < 0:

            # passes 0
            if (current_number != -100) & (prev_number != 0):
                counter += 1
            
            current_number += 100
    
    # right clik
    else:

        # add number
        current_number += calc_number

        # correct for number out of range
        if current_number >= 100:

            # passes 0
            if (current_number != 100) & (prev_number != 0):
                counter += 1

            current_number -= 100
            

    # add to counter when 0
    if current_number == 0:
        counter += 1

    # set new prev_number
    prev_number = current_number

# print result
print(f"Following solution found: {counter}")

```
